```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
api_key = 'RGAPI-ee14068d-a17c-42d4-a146-ad4ab98b03c0'
nbrGames = 20
ig = 'Orval Mfs'
```

```python
puuid = requests.get(f"https://euw1.api.riotgames.com/lol/summoner/v4/summoners/by-name/{ig}?api_key={api_key}").json()["puuid"]
```

```python
games = requests.get(f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count={nbrGames}&api_key={api_key}").json()
```

```python
columns = ['championName', 'win', 'gameDuration', 'teamPosition', 'totalMinionsKilled', 'neutralMinionsKilled', 'totalDamageDealtToChampions', 'goldEarned', 'totalDamageTaken', 'visionScore', 'timeCCingOthers']
```

```python
array = []
```

```python
for i in games:
    json = requests.get(f"https://europe.api.riotgames.com/lol/match/v5/matches/{i}?api_key={api_key}").json()
    index = json['metadata']['participants'].index(puuid)
    player = json['info']['participants'][index]
    player['gameDuration'] = json['info']['gameDuration']
    array.append(list(map(lambda x: player[x], columns)))
```

```python
df = pd.DataFrame(data=array, columns=columns)
```

```python
def formatDf(df):
    df['scoreCreeps'] = df['totalMinionsKilled'] + df['neutralMinionsKilled']
    df['CsPerMinute'] = round(df['scoreCreeps'] / (df['gameDuration'] / 60), 1)
    df['DmgPerMinute'] = (df['totalDamageDealtToChampions'] / (df['gameDuration'] / 60)).astype(int)
    df['DamageTakenPerMinute'] = (df['totalDamageTaken'] / (df['gameDuration'] / 60)).astype(int)
    df.rename(columns={'win': 'winrate'})
    return df
```

```python
dfChamp = formatDf(df)
```

```python
def getMeanFromGames(df, nb, champ):
    return df[df['championName'] == champ.title()][:nb].mean()
```

```python
def getParams(champ):
    def gen(n):
        for i in n:
            yield i

    generator = gen(open('./params.txt', 'r'))
    
    while True:
        str = next(generator)
        if champ in str:
            stats = str.strip().split(';')
            stats.pop(0)
            return {s.split(',')[0]: list(map(float, s.split(',')[1:])) for s in stats}
```

```python
def score(params, data):
    score = 0
    for k, v in params.items():
        score += v[0]*((data[k]-v[1])/(v[2]-v[1]))
    return score
```

```python
def getScore(df, nb, champ):
    data = getMeanFromGames(df, nb, champ)
    params = getParams(champ)
    return score(params, data)
```
