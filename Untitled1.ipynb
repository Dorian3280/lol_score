```python
import requests
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
api_key = 'RGAPI-ee14068d-a17c-42d4-a146-ad4ab98b03c0'
nbrGames = 20
```

```python
puuid = requests.get(f"https://euw1.api.riotgames.com/lol/summoner/v4/summoners/by-name/vayntox?api_key={api_key}").json()["puuid"]
```

```python
games = requests.get(f"https://europe.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count={nbrGames}&api_key={api_key}").json()
```

```python
array = []
columns = ['championName', 'win', 'gameDuration', 'teamPosition', 'totalMinionsKilled', 'neutralMinionsKilled', 'totalDamageDealtToChampions', 'goldEarned']
```

```python
for i in games:
    json = requests.get(f"https://europe.api.riotgames.com/lol/match/v5/matches/{i}?api_key={api_key}").json()
    index = json['metadata']['participants'].index(puuid)
    player = json['info']['participants'][index]
    player['gameDuration'] = json['info']['gameDuration']
    array.append(list(map(lambda x: player[x], columns)))
```

array

```python
df = pd.DataFrame(data=array, columns=columns)
```

```python
df['scoreCreeps'] = df['totalMinionsKilled'] + df['neutralMinionsKilled']
```

```python
df['CsPerMinute'] = round(df['scoreCreeps'] / (df['gameDuration'] / 60), 1)
```

```python
df['DmgPerMinute'] = (df['totalDamageDealtToChampions'] / (df['gameDuration'] / 60)).astype(int)
```

```python
df = df.rename(columns={'win': 'winrate'})
```

```python
vayne = df[df['championName'] == 'Vayne']
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.title("Minions Killed")
plt.scatter(vayne[vayne['winrate'] == True]['winrate'], vayne[vayne['winrate'] == True]['scoreCreeps'], marker="o", c='b', label='win')
plt.scatter(vayne[vayne['winrate'] == False]['winrate'], vayne[vayne['winrate'] == False]['scoreCreeps'], marker="x", c='r', label="lose")
plt.legend()
plt.subplot(1, 2, 2)
plt.title("Winrate")
percent = vayne['winrate'].mean()*100
labels = ['win', 'lose']
plt.pie([percent, 100-percent], labels=labels, autopct='%1.0f%%')
plt.show()
```
